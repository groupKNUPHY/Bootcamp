```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc
from matplotlib import gridspec
```

```python
## PSEUDODATA
w_mu, w_sigma = 80, 80
z_mu, z_sigma = 90, 90
nEvents = 1000000
pseudo_wmt = np.random.normal(w_mu,w_sigma, nEvents)
pseudo_zmt = np.random.normal(z_mu,z_sigma, nEvents)
```

```python
## global size set
plt.rcParams["figure.figsize"] = (8,8)

## grid set
gs = gridspec.GridSpec(2,1,height_ratios=[3,1])

## main plot
plt.subplot(gs[0])
plt.xlim(0,300)
plt.ylabel("Pseudoevents",fontsize = 20)
plt.yscale('log')
a = plt.hist(pseudo_wmt,range=(0,300),bins=10,color='blue',histtype='step',label = 'W')
b = plt.hist(pseudo_zmt,range=(0,300),bins=10,color='red',histtype='step', label = 'Z')
plt.legend(title='Pseudodata set',fontsize = 20)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)

## subplot
plt.subplot(gs[1])
plt.xlim(0,300)
plt.ylim(0.0,2.0)
c = a[0] / b[0]
c = np.nan_to_num(c,posinf=0,neginf=0)
xxx = a[1]
yyy = c
plt.plot(xxx[0:10]+15,yyy,'bo',color='black')
plt.xlabel("M$_{T}$ [GeV]",fontsize = 15)
plt.ylabel("W / Z",fontsize = 15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)

plt.show()
```
